```python
def get_most_frequent(x): 
  freq = np.bincount(x)
  return np.argmax(freq)
```

```python
class points:
  def __init__(self, _cor,_action):
    self.cor = np.array(_cor)
    self.action = _action

  def dist(self, pt):
    d = self.cor-pt.cor
    cal = np.mean(d, 0)
    return np.linalg.norm(d)

class cluster:
  def __init__(self,_points):
    self.points = [pt for pt in _points]
    self.n = len(_points)
  
  def predict(self, input, k=10):
    dist = []
  
    for i, pt in enumerate(self.points):
      dist.append((pt.dist(input),pt.action))
    dist.sort()

    freq = {}
    x = []
    for i in range(k):
      x.append(dist[i][1])
    
    return get_most_frequent(x)
```

```python
def normalize_annotation1(data):
  data = np.array(data)
  mean_pt = np.mean(data,0)
  n_data = (data-mean_pt)
  norm = np.linalg.norm(n_data)
  normal_array = n_data/norm

  return normal_array

def normalize_annotation2(data):
  data = np.array(data)
  mean_pt = np.mean(data,0)
  n_data = (data-mean_pt)

  return n_data

def normalize_annotation3(data):
  data = np.array(data)

  return data
```

```python
one_hand = []
both_hand = []

for fname in train_flist:
  data = train_data[fname]
  for i in range(len(data['annotations'])):
    annotation = np.array(data['annotations'][i]['data']) #change to useful data
    n_annotation = normalize_annotation2(annotation) #normalize 
    pts = points(n_annotation,data['action'][0])
    if len(annotation) == 21:
      one_hand.append(pts)
    else:
      both_hand.append(pts)

print(one_hand)
print(both_hand)
```

```python
oh = []
f_list = []
asdf = []

# Eucledian distance function
def get_dist(plist):
  plist = np.array(plist) # Make plist an np.array list 
  mpt = np.mean(plist,0)
  d = plist-mpt

  dist = []
  for v in d:
    dist.append(np.linalg.norm(v))

  return np.mean(dist)

for fname in train_flist:             # for file name in train_filelist
  data = train_data[fname]            #
  one_hand = len(data['annotations'][0]['data'])==21
  if one_hand:
    f_list.append([])
    asdf.append(fname)
  for i in range(len(data['annotations'])):
    annotation = np.array(data['annotations'][i]['data'])
    n_annotation = normalize_annotation3(annotation)
    n_annotation = n_annotation.reshape(-1)
    if len(annotation) == 21:
      oh.append(n_annotation)
      f_list[-1].append(n_annotation)

print(get_dist(oh))
for i, plist in enumerate(f_list):
  print(asdf[i],get_dist(plist))


```
